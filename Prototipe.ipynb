```python
columns=['Q1','Q3','sample','period','experiment','transition','start','end','ce','name','IS','Group']
```

```python
def SplitInformation(SignalId,columns):
    LColumns=len(columns)
    Information=[]
    IS=0
    for columnp in np.arange(LColumns-2):
        left=SignalId.find(columns[columnp])
        Lleft=len(columns[columnp])+1
        if columnp<LColumns-3:
            right=SignalId.find(columns[columnp+1])
        else:
            right=SignalId.find('_')+1
            name=SignalId[left+Lleft:right-1]
            group=name            
            IS=FindIS(name)
            if IS>0:
                group=DefineGroup(name)
        Information.append(SignalId[left+Lleft:right-1])
    Information.append(IS)
    Information.append(group)
    return Information
```

```python
def FindIS(name):
    IS=0
    if name.find('-d')>0 or name.find('-13C')>0:
        IS=1
    return IS
```

```python
def DefineGroup(name,p=0,IStags=['-d','-13C']):
    ISLoc=name.find(IStags[p])
    if ISLoc>-1:
        group=name[:ISLoc]
        print(group)
    else:
        group=DefineGroup(name,p=p+1)
    return group
```
